```python
%matplotlib notebook
import numpy as np
import scipy as sp
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

N=6  #number of points on the equator
phi=2*np.pi/N  #incremental angle

xgen=np.mgrid[0:N,0:N].reshape(2,-1).T
xgen[4:]=0
print(xgen)
X=np.zeros((N**2,2))
Y=np.zeros((N**2,3))

for p in xgen:
    X[p[0]*N+p[1]]=[p[0]*phi+p[1]*phi/2,np.pi/2-p[1]*phi*np.sin(np.pi/3)]
    Y[p[0]*N+p[1]]=[np.cos(X[p[0]*N+p[1],0])*np.sin(X[p[0]*N+p[1],1]),np.sin(X[p[0]*N+p[1],0])*np.sin(X[p[0]*N+p[1],1]),np.cos(X[p[0]*N+p[1],1])]
#print(X)
print(Y)

fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(Y[:,0],Y[:,1],Y[:,2])
ax.set_box_aspect([1,1,1])
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
```
