```python
import numpy as np
import matplotlib.pyplot as plt

p=0.593
n=1000
X=np.zeros((n,n))
X[1,:]=1
Y=np.random.binomial(1,p,np.shape(X))
print(Y)

def neighbour(X):
  return np.logical_or(np.logical_or(np.roll(X,1,0),np.roll(X,-1,0)),np.logical_or(np.roll(X,1,1),np.roll(X,-1,1)))

def boundary(X):
  return np.logical_and(np.logical_not(X),neighbour(X))

for i in range(1,1000):
  #print(neighbour(X))
  X[np.logical_and(Y,np.logical_and(np.logical_not(X),neighbour(X)))]=i
  X[0,:]=0

fig,ax=plt.subplots((1,1),figsize=(8,8),dpi=200)
ax.imshow(X,cmap='hot')
#plt.imshow(boundary(X),cmap='afmhot')
```
