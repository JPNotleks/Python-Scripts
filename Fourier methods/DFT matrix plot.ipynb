```python
#Plots the DFT matrix with complex domain colouring to reveal the symmetry which the fast Fourier transform exploits.

import numpy as np
import scipy as sp
from scipy import linalg
import matplotlib.pyplot as plt

n=400
X=sp.linalg.dft(n)
X=np.angle(X)
X[X<0]+=2*np.pi
X=X/(2*np.pi)
plt.figure(figsize=(6,6),dpi=150)
plt.imshow(X,cmap='hsv')
plt.show()
print(X)
```
