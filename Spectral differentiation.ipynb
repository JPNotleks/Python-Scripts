```python
import numpy as np
from scipy import fft
from scipy.fftpack import dct
import matplotlib.pyplot as plt

X=np.sin(np.linspace(0,2*np.pi,10,endpoint=False))

plt.plot(np.linspace(0,2*np.pi,10,endpoint=False),X)

Y=fft.fft(X)
print(Y)
Z=np.multiply(Y,np.linspace(0,2*np.pi,10,endpoint=False))

hatu=dct(X,type=2)
N=np.size(X)
#multiply by frequency, minus sign
for i in range(np.size(X)):
    hatu[i]=-(i)*hatu[i]
#shift to left
hatu[0:N-1]=hatu[1:N]
hatu[N-1]=0
Z=hatu

plt.plot(np.linspace(0,2*np.pi,10,endpoint=False),np.abs(Y),color='orange')
plt.plot(np.linspace(0,2*np.pi,10,endpoint=False),np.abs(Z),color='red')
plt.plot(np.linspace(0,2*np.pi,10,endpoint=False),fft.ifft(Z),color='green')
```
