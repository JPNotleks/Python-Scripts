```python
#Solving the heat equation using the implicit scheme
#u_t=mu*u_xx

import numpy as np
import matplotlib.pyplot as plt

n=300 #spatial resolution
m=900 #temporal resolution

mu=1

X=np.zeros((n,m))
X[:,0]=np.sin(np.linspace(0,5*np.pi,n))

A=np.diagflat((1+2*mu)*np.ones((n,1)))-mu*np.diagflat(np.ones((n-1,1)),1)-mu*np.diagflat(np.ones((n-1,1)),-1)
B=np.linalg.inv(A)

for i in range(0,m-1):
  X[:,i+1]=np.matmul(B,X[:,i])
  X[0,i+1]=0; X[-1,i+1]=0

fig=plt.figure(dpi=200,figsize=(15,15))
ax = fig.add_subplot(111)
plt.imshow(X,cmap='inferno')
```
