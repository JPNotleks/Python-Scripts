```python
#A trivial geometric Brownian motion sampler which displays the surprising almost-sure convergence of GBM to zero when
#the volatility is zero. This is highly unintuitive because the marginals of GBM are log-normal and have fat tails. Even
#so, the paths converge to zero almost surely since Brownian motion grows sublinearly. On their way to zero, some paths
#explore significant altitudes before dying out, which explains the fat tails of the marginal laws.

import numpy as np
import matplotlib.pyplot as plt

n=100
m=10000
sigma=0.2
mu=0.0
T=100
dt=T/m

X=np.zeros((n,m))
X[:,0]=1

for i in range(1,m):
  X[:,i]=X[:,i-1]+X[:,i-1]*(mu*dt+np.sqrt(dt)*sigma*np.random.normal(size=(1,n)))

for j in range(1,20):
  plt.plot(np.linspace(0,T,m),(X[j,:]))
```
