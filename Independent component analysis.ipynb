```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA

n=700

s1=np.random.gumbel(loc=0,scale=1,size=(1,n))
s2=np.random.uniform(-5,5,(1,n))
s=np.concatenate((s1,s2),0)

A=np.array([[1,2],[-1,1]])
x=np.matmul(A,s)

fig,ax=plt.subplots(1,3,figsize=(16,8))

transform=FastICA(n_components=2,algorithm='deflation',random_state=0)
y=transform.fit_transform(x)
x2=np.matmul(y,s)
ax[2].scatter(x2[0,:],x2[1,:])

ax[0].scatter(s1,s2)
ax[0].set_xlabel('s1')
ax[0].set_ylabel('s2')
ax[0].set_title('Original s')
ax[1].scatter(x[0,:],x[1,:])
ax[1].set_xlabel('x1')
ax[1].set_ylabel('x2')
ax[1].set_title('Mixed x')
ax[2].set_title('ICA output')
plt.show()
```
