```python
#A custom symplectic fourth-order Runge-Kutta ODE solver used to solve the double pendulum equations and plot them. The animations
#of both the phase-space (3^7 points on a sphere) and the pendulum trajectories are found in the animations folder.

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy
from scipy import integrate
import math
from math import sqrt
import matplotlib.cm as cm
from google.colab import files

x0=np.array([18.6/32*np.pi,5/3*np.pi/2,0,0])
r=0.01
g=9.81
l1=1
l2=1
m1=1
m2=1
N=800

#1000 steps for 1 sec @1000 fps
#new 1000 steps for 0.7 sec @1428 fps
#10000 steps for 10 seconds @ 1000 fps

yeet=0.17560359598
c=np.array([0,2*yeet+1,-4*yeet-1,2*yeet+1])
d=np.array([yeet+1/2,-yeet,-yeet,yeet+1/2])


inc=0


####################################################################################################################################
scale = r
subdiv = 3
name = 'Icosomething'
middle_point_cache = {}

def vertex(x, y, z):
    """ Return vertex coordinates fixed to the unit sphere """

    length = sqrt(x**2 + y**2 + z**2)

    return [(i * scale) / length for i in (x,y,z)]


def middle_point(point_1, point_2):
    """ Find a middle point and project to the unit sphere """

    # We check if we have already cut this edge first
    # to avoid duplicated verts
    smaller_index = min(point_1, point_2)
    greater_index = max(point_1, point_2)

    key = '{0}-{1}'.format(smaller_index, greater_index)

    if key in middle_point_cache:
        return middle_point_cache[key]

    # If it's not in cache, then we can cut it
    vert_1 = verts[point_1]
    vert_2 = verts[point_2]
    middle = [sum(i)/2 for i in zip(vert_1, vert_2)]

    verts.append(vertex(*middle))

    index = len(verts) - 1
    middle_point_cache[key] = index

    return index

PHI = (1 + sqrt(5)) / 2

verts = [
          vertex(-1,  PHI, 0),
          vertex( 1,  PHI, 0),
          vertex(-1, -PHI, 0),
          vertex( 1, -PHI, 0),

          vertex(0, -1, PHI),
          vertex(0,  1, PHI),
          vertex(0, -1, -PHI),
          vertex(0,  1, -PHI),

          vertex( PHI, 0, -1),
          vertex( PHI, 0,  1),
          vertex(-PHI, 0, -1),
          vertex(-PHI, 0,  1),
        ]


faces = [
         # 5 faces around point 0
         [0, 11, 5],
         [0, 5, 1],
         [0, 1, 7],
         [0, 7, 10],
         [0, 10, 11],

         # Adjacent faces
         [1, 5, 9],
         [5, 11, 4],
         [11, 10, 2],
         [10, 7, 6],
         [7, 1, 8],

         # 5 faces around 3
         [3, 9, 4],
         [3, 4, 2],
         [3, 2, 6],
         [3, 6, 8],
         [3, 8, 9],

         # Adjacent faces
         [4, 9, 5],
         [2, 4, 11],
         [6, 2, 10],
         [8, 6, 7],
         [9, 8, 1],
        ]

for i in range(subdiv):
    faces_subdiv = []

    for tri in faces:
        v1 = middle_point(tri[0], tri[1])
        v2 = middle_point(tri[1], tri[2])
        v3 = middle_point(tri[2], tri[0])

        faces_subdiv.append([tri[0], v1, v3])
        faces_subdiv.append([tri[1], v2, v1])
        faces_subdiv.append([tri[2], v3, v2])
        faces_subdiv.append([v1, v2, v3])

    faces = faces_subdiv

################################################################################################################################

verts=np.asarray(verts)
faces=np.asarray(faces)
print(faces[:,0])
    
def dH(t,y):
    t1=y[...,0]
    t2=y[...,1]
    p1=y[...,2]
    p2=y[...,3]
    
    h1=(p1*p2*np.sin(t1-t2))/(
        l1*l2*(m1+m2*np.power(np.sin(t1-t2),2)))
    
    h2=(m2*np.power(l2*p1,2)+(m1+m2)*np.power(l1*p2,2)-2*m2*l1*l2*p1*p2*np.cos(t1-t2))/(
        2*np.power(l1*l2,2)*np.power(m1+m2*np.power(np.sin(t1-t2),2),2))
    
    y1=(l2*p1-l1*p2*np.cos(t1-t2))/(
        np.power(l1,2)*l2*(m1+m2*np.power(np.sin(t1-t2),2)))
    
    y2=(-m2*l2*p1*np.cos(t1-t2)+(m1+m2)*l1*p2)/(
        m2*l1*np.power(l2,2)*(m1+m2*np.power(np.sin(t1-t2),2)))
    
    y3=-(m1+m2)*g*l1*np.sin(t1)-h1+h2*np.sin(2*(t1-t2))
    
    y4=-m2*g*l2*np.sin(t2)+h1-h2*np.sin(2*(t1-t2))
    
    return np.transpose(np.array([y1,y2,y3,y4]))

def RK4(y,dt,tol):
    u=y
    dh=dH(0,u)
    n=1
    dtsub=dt
    norm=math.ceil(np.max(np.abs(dh)))
    if(norm>tol):
        dtsub=dt/norm
        n=norm
    for k in range(n):
        for i in range(4):
            u[2:4]=u[2:4]+c[i]*dtsub*dh[2:4]
            u[0:2]=u[0:2]+d[i]*dtsub*dh[0:2]
    #print(str(n)+' substeps')
    return(u)

#def Verlet(t,y,dt):
#    u=y
#    p12=u[0:2]-dt/2*dH(u)[0:2]
#    q12=u[2:4]+dt/2*(dH([p12,u[2:4]])+dH())


def T(x):
    return np.transpose(np.array([x[...,0],x[...,1],(x[...,2]+x[...,3])/sqrt(2),(-x[...,2]+x[...,3])/sqrt(2)]))
def S(x):
    return np.transpose(np.array([x[...,0],x[...,1],(x[...,2]-x[...,3])/sqrt(2),(x[...,2]+x[...,3])/sqrt(2)]))
    
verts=np.asarray(verts)
faces=np.asarray(faces)

verts4=x0+np.pad(verts,((0,0),(0,1)),'constant')
verts4=S(verts4)

#vertsupload=files.upload()
#verts4=np.load("verts4atTime"+str(inc)+'.npy')


#sul1=np.empty((4,5000))
#sul2=np.empty((4,5000))
#print(odeint(F,verts4[0,:],np.linspace(0,0.1,N)))
#for i in range(1):
#    y1=np.array([np.pi/2,np.pi/3,0,0])
#    y2=np.array([np.pi/2,np.pi/3,0,0])
#    for j in range(5000):
#        y1=RK4(y1,1/1000,1000)
#        sul1[:,j]=y1
#    for j in range(5000):
#        y2=RK4(y2,1/1000,1)
#        sul2[:,j]=y2
#    yTrue1=scipy.integrate.solve_ivp(dH,(0,10),np.array([np.pi/2,np.pi/3,0,0]),method='Radau',t_eval=np.linspace(0,10,1000))
#    yTrue2=scipy.integrate.solve_ivp(dH,(0,10),np.array([np.pi/2,np.pi/3,0,0]),method='Radau',t_eval=np.linspace(0,10,10000))
#    yTrue3=scipy.integrate.solve_ivp(dH,(0,5),np.array([np.pi/2,np.pi/3,0,0]),method='Radau',t_eval=np.linspace(0,5,10000))
#    #print(yTrue)
#    #yAlt=scipy.integrate.solve_ivp(dH,(0,6),verts4[i,:],method='Radau',rtol=1e-6,t_eval=np.linspace(0,6,N)).y
#    #y=scipy.integrate.solve_ivp(F,(0,6),verts4[i,:],t_eval=np.linspace(0,6,N)).y
#    fig,ax=plt.subplots(figsize=(15,15),dpi=1080/15)
#    plt.plot(np.sin(sul1[0,:])+np.sin(sul1[1,:]),-np.cos(sul1[0,:])-np.cos(sul1[1,:]),'orange')
#    plt.plot(np.sin(sul2[0,:])+np.sin(sul2[1,:]),-np.cos(sul2[0,:])-np.cos(sul2[1,:]),'blue')
#    #plt.plot(np.sin(yTrue1.y[0,:])+np.sin(yTrue1.y[1,:]),-np.cos(yTrue1.y[0,:])-np.cos(yTrue1.y[1,:]),'blue')
#    #plt.plot(np.sin(yTrue2.y[0,:])+np.sin(yTrue2.y[1,:]),-np.cos(yTrue2.y[0,:])-np.cos(yTrue2.y[1,:]),'red')
#    plt.plot(np.sin(yTrue3.y[0,:])+np.sin(yTrue3.y[1,:]),-np.cos(yTrue3.y[0,:])-np.cos(yTrue3.y[1,:]),'green')
#    ax.set_xlim((-1.9,-1.6))
#    ax.set_ylim((-0.4,-0.3))
print('done with mesh generation')

n_verts=np.size(verts4[:,0])
print(n_verts)

sol=np.empty((np.size(verts4[:,0]),4,N))
solRK4=np.empty((np.size(verts4[:,0]),4,N))

#print(scipy.integrate.solve_ivp(dH,(inc/100.0,inc/100.0+n/100),np.array([1.82079342 ,2.62650039 ,0. ,        0.        ]),method='Radau',rtol=1e-6,t_eval=np.linspace(0,1,N)).y)

for i in range(np.size(verts4[:,0])):
    #y1=verts4[i,:]
    #print(y1)
    #for j in range(N):
    #    y1=RK4(y1,1/1000,10000)
        #solRK4[i,:,j]=y1
    solRK4[i,:,:]=scipy.integrate.solve_ivp(dH,(inc/100.0*0.7,inc/100.0*0.7+N/100*0.7),verts4[i,:],method='Radau',t_eval=np.linspace(inc/100.0*0.7,inc/100.0*0.7+N/100*0.7,N)).y #odeint(F,verts4[i,:],np.linspace(0,9,N))
    if(i%100==0): print(str(100.0*i/n_verts)+"% ODE")
#print(solRK4[:,:,-1])
#print(solRK4[i,:,0])

#np.save("verts4atTime"+str(inc+N),solRK4[:,:,-1])
#files.download("verts4atTime"+str(inc+N)+".npy")
#print(solRK4[:,:,-1])
print("ODE computation complete")

vertices=verts
T=faces

#print(verts4)
#print(sol[:,1])
#for i in range(N):
#    fig=plt.figure(figsize=(15,15),dpi=1080/15)
#    ax = fig.gca(projection='3d')
#    ax.plot_trisurf(np.sin(sol[:,0,i])+np.sin(sol[:,1,i]), -np.cos(sol[:,0,i])-np.cos(sol[:,1,i]), (np.power(sol[:,0,i],2)+np.power(sol[:,1,i],2))/sqrt(2), triangles = faces, shade=True)
#    #ax.axes.set_xlim3d(left=-2, right=2)
#    #ax.axes.set_ylim3d(bottom=-2, top=2)
#    #ax.axes.set_zlim3d(bottom=0, top=10)
#    plt.savefig("2HamiltonNormie"+str(i),dpi=1080/15)
#    print(i)

#cmmax=0
#for j in range(math.floor(N/1000)):
#    i=1000*j
#    facecol=faces[:,0]
#    print(cm.hot(np.sum(np.power(dH(0,solRK4[:,:,i][facecol,:])[:,0:2],2),axis=1)))
#    facecol=np.sum(np.power(dH(0,solRK4[:,:,i][facecol,:])[:,0:2],2),axis=1)
#    print(facecol)
#    if np.max(facecol)>cmmax: cmmax=np.max(facecol)
  

smear=50
alphas = np.linspace(0.1, 1, smear)
rgba_colors = np.zeros((smear,4))
# for red the first column needs to be one
rgba_colors[:,0] = 173.0/255
rgba_colors[:,1] = 1.0
rgba_colors[:,2] = 47.0/255
# the fourth column needs to be your alphas
rgba_colors[:, 3] = alphas
print(rgba_colors)

for j in range(100,N):
    i=j
#    fig=plt.figure(figsize=(15,15),dpi=1080/15)
#    ax = fig.gca(projection='3d')
    #facecol=faces[:,0]
    #facecol=cm.hot(np.sum(np.power(dH(0,solRK4[:,:,i][facecol,:])[:,0:2],2),axis=1)/1000)
    #print(facecol)
#    ax.plot_trisurf(np.sin(solRK4[:,0,i])+np.sin(solRK4[:,1,i]), -np.cos(solRK4[:,0,i])-np.cos(solRK4[:,1,i]), (np.power(solRK4[:,0,i],2)+np.power(solRK4[:,1,i],2))/sqrt(2), triangles = faces, shade=True)
#    ax.plot_trisurf(solRK4[:,0,i], solRK4[:,1,i], solRK4[:,2,i], triangles=faces, linewidth=0, shade=True)
#    ax.axes.set_xlim3d(left=1.8165, right=1.837)
#    ax.axes.set_ylim3d(bottom=2.606, top=2.629)
#    ax.axes.set_zlim3d(bottom=-0.007, top=0.007)
#    plt.savefig("HamiltonFinalTEST"+str(inc+j),dpi=1080/15)
#    files.download("HamiltonFinalTEST"+str(inc+j)+".png")
#    plt.clf()
#    print(j)
    i=smear
    inc=j
    fig=plt.figure(figsize=(15,15),dpi=1080/15)
    plt.style.use('dark_background')
    ax=fig.add_subplot(111)
    plt.xlim(-2.1,2.1)
    plt.ylim(-2.1,2.1)
    plt.axis('off')
    ax.set_aspect('equal')
    for k in range(n_verts):
        plt.plot(np.append(np.sin(solRK4[k,0,inc]),0),np.append(-np.cos(solRK4[k,0,inc]),0),color=(0,32./255,194./255),linewidth=1)
        plt.plot(np.append(np.sin(solRK4[k,0,inc]),np.sin(solRK4[k,0,inc])+np.sin(solRK4[k,1,inc])), np.append(-np.cos(solRK4[k,0,inc]),-np.cos(solRK4[k,0,inc])-np.cos(solRK4[k,1,inc])),color=(9./255,65./255,194./255),linewidth=1)
    for k in range(n_verts):
        for u in range(smear):
            plt.plot(np.append(np.sin(solRK4[k,0,inc-u-1])+np.sin(solRK4[k,1,inc-u-1]),np.sin(solRK4[k,0,inc-u])+np.sin(solRK4[k,1,inc-u])), np.append(-np.cos(solRK4[k,0,inc-u-1])-np.cos(solRK4[k,1,inc-u-1]),-np.cos(solRK4[k,0,inc-u])-np.cos(solRK4[k,1,inc-u])),color=(173./255,1,47./255),linewidth=1,alpha=np.power(1-u/smear,3))
    plt.savefig("Hamiltonpath"+str(j),dpi=1080/15)
    print(j)
    #files.download("Hamiltonpath"+str(j)+".png")
    plt.clf()
```
