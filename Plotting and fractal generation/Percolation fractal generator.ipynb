```python
#Percolation refers to the flow of liquid through a granular medium like a sponge or sand. When the permissivity of the medium is
#just right (59.3% of space is obstruced), the percolation is critical and forms a fractal boundary with a Hausdorff dimension of 1.5
#This script animates the percolation as it progresses, and extracts the boundary once is stops flowing. Animations and boundary images
#are found in the animations folder.

import numpy as np
from matplotlib import pyplot as plt
from google.colab import drive
from google.colab import files
import os

!pip install -U -q PyDrive
#pip install -U -q PyDrive
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

drive.mount("/content/drive/percolationimages")

n=1000
p=0.593

X=np.zeros((n,n),dtype=int)
Y=np.random.binomial(1,p,np.shape(X))
np.save("percolationmatrix",Y)
touched=np.zeros(np.shape(X),dtype=int)
col=np.zeros(np.shape(X))
X[2,:]=1
xmin=0;xmax=5;ymin=n//2-1;ymax=n//2+1

def neighbour(X):
    return np.logical_or(np.logical_or(np.roll(X,1,0),np.roll(X,-1,0)),np.logical_or(np.roll(X,1,1),np.roll(X,-1,1)))

cont=1
counter=0

for it in range(1,1000):
    X[np.logical_and(np.logical_not(X),np.logical_and(np.logical_and(np.logical_not(touched),neighbour(X)),Y))]=it
    X[0,:]=0
    if(it%13==0):
        #fig, axs = plt.subplots(1, 1,figsize=(10.8, 10.8), dpi=200)
        #plt.axis('off')
        #axs.imshow(X/np.max(X),cmap='twilight')
        #plt.savefig("perc9image"+str(it/13)+".png",dpi=200)
        #plt.close()
        
        #plt.savefig("sgd"+str(int(i/10))+".png",bbox_inches='tight')
        ###files.download("sgd"+str(i/50)+".png")
        #uploaded = drive.CreateFile({'title': 'sgd'+str(int(i/10))+'.png'})
        #uploaded.SetContentFile('sgd'+str(int(i/10))+'.png')
        #uploaded.Upload()
        print(it)
print("done")
np.save("Xperc",X)
#fig, axs = plt.subplots(1, 1,figsize=(10.8, 10.8), dpi=1000)
#plt.axis('off')
#axs.imshow(X/np.max(X),cmap='twilight')
#plt.savefig("perc9imagefinal.png",dpi=1000)

C=np.zeros(np.shape(X))
C[-1,:]=-1
for it in range(int(n*1.5)):
    C[np.logical_and(np.equal(C,0),np.logical_and(neighbour(C),np.equal(X,0)))]=-1
    if(it%100==0):
        print(it)

C[np.logical_and(np.equal(C,-1),neighbour(X))]=1
C[np.equal(C,-1)]=0
C[0,:]=0
C[1,:]=0
touchedC=C

for it in range(1,70):
    C[np.logical_and(np.logical_not(C),np.logical_and(np.logical_not(touchedC),neighbour(C)))]=np.power(0.9,it)
    C[0,:]=0
    print(it)
print("done")

np.save("finalbdy",C)

fig, axs = plt.subplots(1, 1,figsize=(10.8, 10.8), dpi=100)
plt.axis('off')
axs.imshow(C,cmap='hot')
#plt.savefig("perc9bdy.png",dpi=1000)
plt.savefig("per0.png",bbox_inches='tight')
###files.download("sgd"+str(i/50)+".png")
uploaded = drive.CreateFile({'title': 'per0.png'})
uploaded.SetContentFile('per0.png')
uploaded.Upload()

#for i in range(200):
#    fig, axs = plt.subplots(1, 1,figsize=(10.8, 10.8), dpi=200)
#    plt.axis('off')
#    Y=X/np.max(X)
#    xmin=n//2-int(n//2*np.exp(-i/70))
#    xmax=n//2+int(n//2*np.exp(-i/70))
#    ymin=n//2-int(n//2*np.exp(-i/70))
#    ymax=n//2+int(n//2*np.exp(-i/70))
#    axs.imshow(Y[xmin:xmax,ymin:ymax],cmap='twilight',norm = colors.Normalize(vmin=0, vmax=1))
#    plt.savefig("perc9image"+str(i+550)+"png",dpi=200)
```
