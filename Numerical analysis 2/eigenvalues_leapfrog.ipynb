```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

a = 0.0
b = 6.0
N = 100

I = np.linspace(a, b, N)

eig1_exa = np.ndarray(N, dtype=np.complex128)
eig2_exa = np.ndarray(N, dtype=np.complex128)

eig1_app = np.ndarray(N, dtype=np.complex128)
eig2_app = np.ndarray(N, dtype=np.complex128)


# COMPUTE THE EXACT EIGENVALUES
# functions
def eig1_exa_fun(s):
    return np.exp(1j*s)

def eig2_exa_fun(s):
    return np.exp(-1j*s)

# vectors containing exact eigenvalues
for i in range(N):
    eig1_exa[i] = eig1_exa_fun(I[i])
    eig2_exa[i] = eig2_exa_fun(I[i])
    

# COMPUTE THE APPROXIMATE EIGENVALUES
# functions
def eig1_app_fun(s): 
    return 1/2*(2-np.power(s,2)+cmath.sqrt(np.power(s,4)-4*np.power(s,2)))

def eig2_app_fun(s):
    return 1/2*(2-np.power(s,2)-cmath.sqrt(np.power(s,4)-4*np.power(s,2)))

# vectors containing approximate eigenvalues
for i in range(N):
    eig1_app[i] = eig1_app_fun(I[i])
    eig2_app[i] = eig2_app_fun(I[i])

# PLOT EXACT VS APPROXIMATE EIGENVALUES
plt.plot(I,np.real(eig1_exa),'g-',label="Re_eig1_exact") 
plt.plot(I,np.real(eig1_app),'m+',label="Re_eig1_approx")
plt.xlabel("$\Delta t\omega$"),plt.ylabel("Real part")
plt.legend(loc="upper right"),plt.savefig("real1.pdf")

plt.figure()
plt.plot(I,np.real(eig2_exa),'g-',label="Re_eig2_exact") 
plt.plot(I,np.real(eig2_app),'m+',label="Re_eig2_approx")
plt.xlabel("$\Delta t\omega$"),plt.ylabel("Real part")
plt.legend(loc="upper right"),plt.savefig("real2.pdf")

plt.figure()
plt.plot(I,np.imag(eig1_exa),label="Im_eig1_exact") 
plt.plot(I,np.imag(eig1_app),'+',label="Im_eig1_approx")
plt.xlabel("$\Delta t\omega$"),plt.ylabel("Imaginary part")
plt.legend(loc="upper right"),plt.savefig("imag1.pdf")

plt.figure()
plt.plot(I,np.imag(eig2_exa),label="Im_eig2_exact") 
plt.plot(I,np.imag(eig2_app),'+',label="Im_eig2_approx")
plt.xlabel("$\Delta t\omega$"),plt.ylabel("Imaginary part")
plt.legend(loc="upper right"),plt.savefig("imag2.pdf")
```
