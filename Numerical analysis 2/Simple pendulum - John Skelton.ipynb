```python
%matplotlib inline

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def pend(y,t,g,l):
    return [y[1],-g/l*np.sin(y[0])]

t=np.linspace(0,10,101)
y=odeint(pend,[np.pi/4,0],t,args=(9.81007,1))

fig=plt.figure()
plt.plot(t,y[:,0],label='theta(t)')

def harmonicApproxilator(y,t,g,l):
    return [y[1],g/l*y[0]]

y2=odeint(harmonicApproxilator,[-np.pi/4,0],t,args=(9.81007,1))
plt.plot(t[0:10],y2[:,0][0:10]+np.pi,label='theta(t) badly linearised')

plt.yticks(np.linspace(-np.pi,np.pi,9,endpoint=True))
plt.legend()

#The linearisation of the equation fails as we approach the stable equilibrium point theta=0, because it is nonzero there.
```
