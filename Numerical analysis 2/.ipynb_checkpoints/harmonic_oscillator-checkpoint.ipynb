```python

import numpy as np
import matplotlib.pyplot as plt

T = np.asarray([0,10]) #### Start and end times
y0 = np.asarray([1,0]) #### Initial Value

# right hand side
def f(t,y):
    return np.asarray([-y[:][1],y[:][0]])

def improved_euler(f,y0,T,h):
    # Initialize variables
    N = int(np.ceil(T[1]-T[0])/h)# number of steps
    h = (T[1]-T[0])/N # step size 
    d = len(y0) # dimension of solution
    t = np.linspace(T[0],T[1],N+1) # time grid
    y = np.zeros((d,N+1)) # solution
    y[:,0] = y0 # s e t initial value
    
    for j in range(0,N):
        y[:,j+1] = y[:,j]+1/2*h*(f(t,y[:,j])+f(t,y[:,j]+h*f(t,y[:,j])))
         
    return np.array([t,y])

# Call the improved Euler method function here
[t,u] = improved_euler(f,y0,T,0.05)

# Plot
plt.plot(t,u[0,:],label="x")
plt.plot(t,u[1,:],label="dx/dt")
plt.xlabel('t')
plt.legend(loc="upper right")
```
