```python
# dt(u)=cddx(u)

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as wg
import seaborn as sb
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure()

n=30              #space resolution
tMax=10           #time resolution
c=0.1             #conductivity
k=0.01             #time step
j=tMax*int(1/k)   #number of t iterations
h=np.pi/(n-1)        #space step
r=c*k/h**2

x=np.linspace(0,np.pi,n)              #x domain
T=np.linspace(0,tMax,j)   #t domain

U=np.zeros((j,n-2))       #Matrix of space and time values
U[0]=np.sin((x[1:-1])**2)

M=np.zeros((n-2,n-2))                 #Tridiagonal matrix of U(t+1)
M[0,0]=2+2*r;M[0,1]=-r
M[n-3,n-4]=-r;M[n-3,n-3]=2+2*r

for i in range(1,n-3):
    M[i,i-1]=-r
    M[i,i]=2+2*r
    M[i,i+1]=-r
    
N=np.zeros((n-2,n-2))                 #Tridiagonal matrix of U(t)
N[0,0]=2-2*r;N[0,1]=r
N[n-3,n-4]=r;N[n-3,n-3]=2-2*r

for i in range(1,n-3):
    N[i,i-1]=r
    N[i,i]=2-2*r
    N[i,i+1]=r

#print(M)
#print(N)

for i in range(1,j):
    U[i]=np.linalg.solve(M,np.dot(N,U[i-1]))
#print(U)

x,T=np.meshgrid(x,T)

z=np.zeros((j,1))       #regenerate boundaries
U=np.append(z,U,axis=1)
U=np.append(U,z,axis=1)

ax=fig.add_subplot(111, projection='3d')
#print(U)

ax.plot_wireframe(T,x,U,color='red')
#ax.plot_wireframe(T,x,np.sin(x)*np.exp(-c*T),color='green')

def f(x,t):
    return np.sin(x)*np.exp(-t)

#def plotting(t):
#    plt.plot(x,f(x,t))
#    plt.grid(True)
#    major_xticks=np.linspace(0,np.pi,5)
#    plt.xticks(major_xticks)
#    major_yticks=np.linspace(0,1,5)
#    plt.yticks(major_yticks)
#    plt.xlim((0,np.pi))
#    plt.ylim((0,1))
    
#wg.interact(plotting,t=(0,5))
```
