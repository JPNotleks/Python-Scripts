```python
#2D heat equation using explicit method
%matplotlib notebook
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
#import matplotlib.animation as animation
np.set_printoptions(precision=2)
import time

v=0.1            #propagation speed (m/s)
n=5          #spatial points x
m=5          #spatial points y
l=100         #temporal points
[x1,x2]=[0,np.pi]    #spatial boundaries x (m)
[y1,y2]=[0,np.pi]    #spatial boundaries y (m)
[t1,t2]=[0,10]  #temporal boundaries (s)
dx=(x2-x1)/(n-1)     #spatial step size (m)
dy=(y2-y1)/(m-1)
dt=(t2-t1)/(l-1)   #temporal step size (s)
x=np.linspace(x1,x2,n)
y=np.linspace(y1,y2,m)
t=np.linspace(t1,t2,l)

a=(dx**2+dy**2)/(dx**2*dy**2)
#a=(v*dt/dx)**2      #alpha^2
#print(a)
#if a>1:
#    print(a)
#    raise Exception("a>1, might encounter numerical instability")

u=np.zeros((l,n,m))  #the grid in t,x,
X,Y=np.meshgrid(x,y)
fx=np.sin(X)#+np.sin(2*x)  #initial u(t=0)
fy=np.sin(Y)
f=fx*fy
gx=0*X    #initial u'(t=0)
gy=0*Y
g=gx+gy

fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.set_zlim(-1, 1)
#ax.plot_wireframe(X,Y,f,color='blue')

plt.show()

u[0,:,:]=f
L0=1/dy**2*(u[0,1:n-1,2:m]+u[0,1:n-1,0:m-2]) + 1/dx**2*(u[0,2:n,1:m-1]+u[0,0:n-2,1:m-1]) -2*a*u[0,1:n-1,1:m-1]
u[1,1:n-1,1:m-1]=v**2/2*dy**2*L0 + u[0,1:n-1,1:m-1] + dy*g[1:n-1,1:m-1]

for k in range(1,l-1):
    L=1/dy**2*(u[k,1:n-1,2:m]+u[k,1:n-1,0:m-2]) + 1/dx**2*(u[k,2:n,1:m-1]+u[k,0:n-2,1:m-1]) -2*a*u[k,1:n-1,1:m-1]    #Laplacian at time k of sheet x,y
    u[k+1,1:n-1,1:m-1]=v**2*dy**2*L + 2*u[k,1:n-1,1:m-1] - u[k-1,1:n-1,1:m-1]

plt.show()

print(np.round(u,2))

#global wframe
#def animate(i):
#    #self.subplot.cla()
#    global wframe
#    if wframe:
#        ax.collections.remove(wframe)
#    wframe=ax.plot_wireframe(X,Y,u[i,:,:],color='blue')

wframe=None
for i in range(0,100):
    # If a line collection is already remove it before drawing.
    if wframe:
        ax.collections.remove(wframe)

    # Plot the new wireframe and pause briefly before continuing.
    wframe=ax.plot_wireframe(X,Y,u[i,:,:],color='blue')
    plt.pause(.1)

#ani = animation.FuncAnimation(fig, animate, interval=100,frames=l)

plt.show()
```
